# Comparison

This notebook contains code to generate results for downstream tasks except for self-storage

```python
import io as b_io
import os
from pathlib import Path
import dill

from mosaiks import config as c
from mosaiks import transforms
from mosaiks.utils.imports import *
from mosaiks.utils import io as mio
```

We use ridge regression for all comparisons

```python
solver = solve.ridge_regression
```

This cell loads the embeddings generated from our multitask model

```python
X_multitask = {}
latlons_multitask = {}
sample = "POP"

features_path = Path(c.features_dir) / f"resnet18multitask_{sample}.pkl"
print(features_path)
with open(features_path, "rb") as f:
    data = np.load(f, allow_pickle=True)
temp = b_io.BytesIO(data["X"])
temp2 = np.load(temp, allow_pickle=True)
features = temp2
latlons_samp = data["latlon"]
ids_x = data["ids_X"].astype('str')

X_multitask[sample] = pd.DataFrame(features, index=ids_x)
latlons_multitask[sample] = pd.DataFrame(latlons_samp, index=ids_x, columns=["lat", "lon"])

display(X_multitask[sample])
display(latlons_multitask[sample])
```

```python
labels_to_run = ["nightlights"]
```

This cell loads the embeddings generated from all ResNet18s, and combines them. Note we actually adding features from a ResNet18 trained on a task if we will currently train on that task.

```python
tasks_UAR = ["treecover", "elevation", "population",]
tasks_POP = ["nightlights", "income", "roads", "housing",]

X_combined = {}
latlons_combined = {}
sample = "POP"

features_path = Path(c.features_dir) / f"resnet18_combined_{sample}.pkl"
print(features_path)
with open(features_path, "rb") as f:
    data = np.load(f, allow_pickle=True)
X_list = []
for task in tasks_UAR + tasks_POP:
    if task == "nightlights":
        continue
    temp = b_io.BytesIO(data[f"X_{task}"]["X"])
    temp2 = np.load(temp, allow_pickle=True)
    X_list.append(temp2)
features = np.concatenate(X_list, axis=1)

latlons_samp = data["latlon"]
ids_x = data["ids_X"].astype('str')

X_combined[sample] = pd.DataFrame(features, index=ids_x)
latlons_combined[sample] = pd.DataFrame(latlons_samp, index=ids_x, columns=["lat", "lon"])

display(X_combined[sample])
display(latlons_combined[sample])
```

```python
# This code is based off of the ridge regression from the MOSAIKS codebase.

label = labels_to_run[0]

print("*** Running regressions for: {}".format(label))

## Set some label-specific variables
this_cfg = io.get_filepaths(c, label)
c_app = getattr(this_cfg, label)
outcome_name = c_app["variable"]
sampling_type = c_app["sampling"]  # UAR or POP

best_lambda_fpath = join(c.fig_dir_sec, "best_lambda_tl.npz")
lambdas = io.get_lambdas(c, label, best_lambda_fpath=None)

if c_app["logged"]:
    bounds = np.array([c_app["us_bounds_log_pred"]])
else:
    bounds = np.array([c_app["us_bounds_pred"]])

# Set solver arguments
solver_kwargs = {
    "lambdas": lambdas,
    "return_preds": True,
    "svd_solve": False,
    "clip_bounds": bounds,
}
if label in ["income", "roads"]:
    solver_kwargs["lambdas"] = np.logspace(-3, 4, 9)

## get X, Y, latlon values of training data
(
    this_X,
    this_X_test,
    this_Y,
    this_Y_test,
    latlons_train,
    _,
) = parse.merge_dropna_transform_split_train_test(
    this_cfg, label, X_combined[sampling_type], latlons_combined[sampling_type]
)

## Train model using ridge regression and 5-fold cross-valiation
## (number of folds can be adjusted using the argument n_folds)
print("Training model...")
kfold_results = solve.kfold_solve(
    this_X,
    this_Y,
    solve_function=solver,
    num_folds=this_cfg.ml_model["n_folds"],
    return_model=True,
    **solver_kwargs
)
print("")

## Store the metrics and the predictions from the best performing model
best_lambda_idx, _, _ = ir.interpret_kfold_results(
    kfold_results, "r2_score", hps=[("lambdas", solver_kwargs["lambdas"])]
)
best_lambdas = np.array(
    [solver_kwargs["lambdas"][np.asarray(best_lambda_idx).squeeze()]]
)

# save best lambdas
np.savez(best_lambda_fpath, best_lambda=best_lambdas)

## Retrain a model using this best lambda:
holdout_results = solve.single_solve(
    this_X,
    this_X_test,
    this_Y,
    this_Y_test,
    lambdas=best_lambdas,
    return_preds=True,
    return_model=False,
    clip_bounds=bounds,
    svd_solve=False,
)
```

```python
plots.performance_density(
    holdout_results,
    "Ridge",
    val=outcome_name,
    save_dir=None,
    app_name=label,
    kind="scatter",
    size=5
)
```
